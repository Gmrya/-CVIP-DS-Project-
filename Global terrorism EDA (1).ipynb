# Exploratory data analysis - Global Terrorism Dataset



### About dataset:
The Global Terrorism Database (GTD) is an open-source database including information on terrorist attacks around the world from 1970 through 2017. The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland.

### Objective:

To perform Exploratory Data Analysis (EDA) on the terrorism dataset in order to extract meaningful insights and patterns. By analyzing the dataset, we aim to uncover trends, identify high-risk regions, understand attack characteristics, and potentially reveal factors that contribute to terrorist activities.

## Data preprocessing

```python
missing_percentage=terrorism_data.isna().sum()/len(terrorism_data)*100
missing_percentage
```

```python
limit= 50
```

```python
dropcolumns = missing_percentage[missing_percentage > limit].index.tolist()
```

```python
dropcolumns
```

```python
terrorism_data.drop(columns=dropcolumns, inplace= True, errors='ignore')
```

```python
terrorism_data.columns
```

```python
terrorism_data.shape 
```

```python
terrorism_data.isna().sum().sort_values(ascending= False)/len(terrorism_data)*100
```

```python
terrorism_data.rename(columns={'iyear':'Year', 'imonth': 'Month', 'iday': 'Day' , 'nwound' :'Wounded' , 'nkill':'Killed' , 'weaptype1_txt':'Weapontype' ,'attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','region_txt':'Region','country_txt':'Country'}, inplace= True)
```

```python
terrorism_data = terrorism_data[['Year','Month','Day','Country','Region','Wounded', 'Killed','Weapontype','Attacktype','Targettype','gname']]
terrorism_data
```

```python
terrorism_data.isna().apply(pd.value_counts)
```

### Number of Attacks by Region

```python
f = plt.figure(figsize=(20, 10))
sns.countplot(y='Region', data=terrorism_data)
plt.ylabel('Region', fontsize=14)
plt.xlabel('Number of attacks', fontsize=14)
plt.title('Terrorist Attacks by Region ', fontsize=14)
```

### Number of Attacks by Year

```python
f = plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=terrorism_data)
plt.ylabel('Number of attacks', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.xticks(rotation=45)
plt.title('Terrorist Attacks by Year', fontsize=14)
```

### Number of attacks per attack type

```python
f= plt.figure(figsize=(20, 10))  
sns.countplot(x='Attacktype',data=terrorism_data)
plt.xlabel('Type of Attack',fontsize=14)
plt.ylabel('Counts',fontsize=14)
plt.title('Number of attacks per attack type', fontsize=14)  
```

### Number of attacks by country(Top 15)

```python
fig= plt.figure(figsize=(20, 10))
sns.barplot(x=terrorism_data['Country'].value_counts()[0:15].index, y=terrorism_data['Country'].value_counts()[0:15])
plt.xticks(rotation=45)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of attacks', fontsize=14)
plt.title('Top 15 most attacked countries', fontsize=14)
      
```

### Number of attacks per targettype

```python
f= plt.figure(figsize=(25,10))  
sns.countplot(x='Targettype',data=terrorism_data)
plt.xticks(rotation=90)
plt.xlabel('Target',fontsize=14)
plt.ylabel('Counts',fontsize=14)
plt.xticks(rotation=45)
plt.title('Number of attacks per target type', fontsize=12)  
```

###  Terror attacks by region in each year

```python
pd.crosstab(terrorism_data.Year,terrorism_data.Region).plot(kind='area',stacked=False,figsize=(25,10)) 
plt.title('Terror attacks by region in each year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

### Terror attacks by Weapontype in each year

```python
fig = plt.figure(figsize=(16, 10))
pd.crosstab(terrorism_data.Year,terrorism_data.Weapontype).plot(figsize=(14, 10), fontsize=13)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Attacks', fontsize=12)
plt.title('Number of terror attacks per Weapontype in each year', fontsize=12)   
```

### Top 15 most active terrorist groups

```python
plt.figure(figsize=(14,7))
sns.barplot(x=terrorism_data['gname'].value_counts()[1:15].values,y=terrorism_data['gname'].value_counts()[1:15].index)
plt.title('Most Active Terrorist Groups')
plt.ylabel('Name of Group')
plt.xlabel('No. of Attacks')
```
